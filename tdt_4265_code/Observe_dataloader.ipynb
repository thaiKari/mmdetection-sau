```python
from dataloaders import load_cifar10
import matplotlib.pyplot as plt
```

```python
Batch_size = 4
```

```python
dataloader_train, dataloader_val, dataloader = load_cifar10(Batch_size)
```

```python
CLASSES=[
    
'airplane										',
'automobile										',
'bird										',
'cat										',
'deer										',
'dog										',
'frog										',
'horse										',
'ship										',
'truck'
]
```

```python
for i_batch, (X_batch, Y_batch) in enumerate(dataloader):
    print(i_batch, X_batch.size(), Y_batch.size())
    
    for i in range(Batch_size):
        print('label',Y_batch[i].item(),  CLASSES[Y_batch[i].item()])
        im_i = X_batch[i,:,:,:]
        #print(im_i.shape)
        plt.figure()
        plt.imshow(im_i[0,:,:])
        plt.show()
        
    break
        
        
        
```
